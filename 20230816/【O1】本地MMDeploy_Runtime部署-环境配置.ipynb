## 配置环境

```shell

conda install nb_conda_kernels -y

conda create -n mmseg python=3.8 -y

conda activate mmseg

conda install ipykernel -y

ipython kernel install --user --name=mmseg

jupyter notebook
```

## 打开每一个notebook后，都需要切换kernel为`mmseg`虚拟环境
